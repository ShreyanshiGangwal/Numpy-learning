# Basic NumPy

```python
import numpy as np
```

 ---------------------------------------------------
### 1. Create a NumPy array from a Python list

```python
list = [1, 2, 3, 4, 5]
arr = np.array(list)

print('Numpy Array:', arr)
```

 ---------------------------------------------------
### 2. Create: 1D array, 2D array & 3D array.

```python
# 1D Array...
arr1 = np.array([1, 2, 3, 4, 5])
print('1D Array:', arr1)

# 2D Array...
arr2 = np.array([[1, 2, 3],[4, 5, 6]])
print('\n2D Array:')
print(arr2)

# 3D Array...
arr3 = np.array([[[1, 2], [3, 4]],   
                 [[5, 6], [7,8]]])
print('\n3D Array:')
print(arr3)
```

 ---------------------------------------------------
### 3. Find: shape, size, ndim, dtype.

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print('Array:\n', arr)

# Array properties...
print('\nShape of the array:', arr.shape)   # shape
print('\nSize of the array:', arr.size)   # size
print('\nNumber of dimensions:', arr.ndim)   # ndim
print('\nData type of the array:', arr.dtype)   # dtype
```

 ---------------------------------------------------
### 4. Create arrays using: zeros(), ones(), full(), arange(), linspace(), eye().

```python
# Creating arrays with specific values...
zeros_array = np.zeros((2, 3))
print('Array of zeros:\n', zeros_array)

ones_array = np.ones((2, 3))
print('\nArray of ones:\n', ones_array)

full_array = np.full((2, 3), 7)
print('\nArray filled with 7:\n', full_array)

arange_array = np.arange(0, 10, 2)
print('\nArray with arange:\n', arange_array)

linspace_array = np.linspace(0, 1, 5)
print('\nArray with linspace:\n', linspace_array)

identity_matrix = np.eye(3)
print('\nIdentity Matrix:\n', identity_matrix)
```

 ---------------------------------------------------
### 6. Convert list → array and array → list.

```python
# List to Array...
list = [10, 20, 30, 40]
arr = np.array(list)

print('Converted Array:', arr)

# Array to List...
arr_to_list = arr.tolist()

print('\nConverted List:', arr_to_list)
```

 ---------------------------------------------------
### 7. Create array of even numbers from 1–100.

```python
even_arr = np.arange(2, 101, 2)

print('Array of Even Numbers:', even_arr)
```

 ---------------------------------------------------
### 8. Create float datatype array.

```python
float_array = np.array([1, 2, 3, 4, 5], dtype=float)
print('Float DataType Array:', float_array)
print('\nData type of the array:', float_array.dtype)
```

 ---------------------------------------------------
### 9. Create boolean array.

```python
bool_array = np.array([True, False, True, False])
print('Boolean Array:', bool_array)
print('\nData type of the boolean array:', bool_array.dtype)
```

 ---------------------------------------------------
### 10. Compare memory usage of list vs NumPy array.

```python
import sys

# Python List...
my_list = [1, 2, 3, 4, 5]

# NumPy Array...
my_array = np.array([1, 2, 3, 4, 5])

# Memory Usage...
print('Memory usage by Python list:', sys.getsizeof(my_list), 'bytes')
print('\nMemory usage by NumPy array:', my_array.nbytes, 'bytes')
```
