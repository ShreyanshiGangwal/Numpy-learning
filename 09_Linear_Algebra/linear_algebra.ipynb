# Linear Algebra

```python
import numpy as np
```

 ---------------------------------------------------
### 1. Matrix multiplication

```python
A = np.array([
    [1, 2],
    [3, 4]
])

B = np.array([
    [5, 6],
    [7, 8]
])

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)

matrix_multi = np.matmul(A, B)

print("\nMatrix Multiplication:\n", matrix_multi)
```

 ---------------------------------------------------
### 2. Dot product

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

dot_product = np.dot(arr1, arr2)

print("Array 1:", arr1)
print("Array 2:", arr2)

print("\nDot Product:", dot_product)
```

 ---------------------------------------------------
### 3. Cross product

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

cross_product = np.cross(arr1, arr2)

print("Array 1:", arr1)
print("Array 2:", arr2)

print("\nCross Product:", cross_product)
```

 ---------------------------------------------------
### 4. Transpose matrix

```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

transpose = matrix.T

print("Matrix:\n", matrix)

print("\nTranspose Matrix:\n", transpose)
```

 ---------------------------------------------------
### 5. Find determinant
- Central Formula:
- det(A) = ad - bc

```python
matrix = np.array([
    [4, 6],
    [3, 8]
])

print("Matrix:\n", matrix)

determinant = np.linalg.det(matrix)

print("\nDeterminant:", determinant)

# Formula Visualization
# det(A) = ad - bc
# :contentReference[oaicite:0]{index=0}
```
