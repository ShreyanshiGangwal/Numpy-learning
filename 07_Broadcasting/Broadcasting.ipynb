# Broadcasting in NumPy

```python
import numpy as np
```

 ---------------------------------------------------
### 1. Add scalar to matrix

```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print('Matrix:\n', matrix)

add = matrix + 10

print("\nAfter Adding Scalar:\n", add)
```

 ---------------------------------------------------
### 2. Add vector to matrix rows

```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("Matrix:\n", matrix)

row_vector = np.array([10, 20, 30])

print("\nRow Vector:", row_vector)

result = matrix + row_vector

print("\nAfter Row-wise Addition:\n", result)
```

 ---------------------------------------------------
### 3. Add vector to matrix columns

```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print('Matrix:\n', matrix)

column_vector = np.array([[10],
                          [20]])
print('\nColumn Vector:\n', column_vector)

result = matrix + column_vector

print('\nAfter Column-wise Addition:\n', result)
```

 ---------------------------------------------------
### 4. Multiply arrays of different shapes

```python
arr1 = np.array([
    [1],
    [2],
    [3]
])

arr2 = np.array([10, 20, 30])

result = arr1 * arr2

print("First Array:\n", arr1)
print("\nSecond Array:\n", arr2)

print("\nMultiplication Result:\n", result)
```

 ---------------------------------------------------
### 5. Solve shape mismatch problem

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([[10],
                 [20],
                 [30]])

# --- Using broadcasting ---
result = arr1 + arr2

print("First Array:", arr1)
print("\nSecond Array:\n", arr2)

print("\nAfter Solving Shape Mismatch:\n", result)
```

 ---------------------------------------------------
### 6. Create multiplication table using broadcasting

```python
numbers = np.arange(1, 11)

table = numbers[:, np.newaxis] * numbers

print("Multiplication Table:\n")
print(table)
```

 ---------------------------------------------------
### 7. Normalize matrix using broadcasting
Formula:
- (x - min) / (max - min)

```python
matrix = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print('Matrix:')
print(matrix)

norm = (matrix - matrix.min()) / (matrix.max() - matrix.min())

print('\nNormalized Matrix:')
print(norm)
```

 ---------------------------------------------------
### 8. Apply row-wise operations
- Add row sums to each row

```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("Matrix:\n", matrix)

row_sum = matrix.sum(axis=1, keepdims=True)

print("\nRow Sums:\n", row_sum)

result = matrix + row_sum

print("\nAfter Row-wise Operation:\n", result)
```

 ---------------------------------------------------
### 9. Apply column-wise operations
- Add column sums to each column

```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("Matrix:\n", matrix)

column_sum = matrix.sum(axis=0)

print("\nColumn Sums:", column_sum)

res = matrix + column_sum

print("\nAfter Column-wise Operation:\n", result)
```

 ---------------------------------------------------
### 10. Create distance matrix
- Distance = |x - y|

```python
points = np.array([1, 3, 5, 7])

distance_matrix = np.abs(points[:, np.newaxis] - points)

print("Points:", points)

print("\nDistance Matrix:")
print(distance_matrix)
```
