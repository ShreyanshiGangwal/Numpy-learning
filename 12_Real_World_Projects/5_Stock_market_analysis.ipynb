# Stock Market Mini Analysis
---

```python
import numpy as np
```

### Take Stock Prices

```python
days = int(input("Enter number of trading days: "))

prices = []

for i in range(days):

    price = float(input(f"Enter stock price for Day {i+1}: "))

    prices.append(price)

prices = np.array(prices)

print("Stock Prices:\n")
print(prices)
```

### Basic Statistics

```python
print("----- STOCK ANALYSIS -----\n")

print("Average Price:",
      np.mean(prices))

print("Highest Price:",
      np.max(prices))

print("Lowest Price:",
      np.min(prices))

print("Price Variance:",
      np.var(prices))

print("Standard Deviation:",
      np.std(prices))
```

### Daily Price Change

```python
daily_change = np.diff(prices)

print("Daily Price Changes:\n")
print(daily_change)
```

### Profit or Loss

```python
print("----- DAILY STATUS -----\n")

for i in range(len(daily_change)):

    if daily_change[i] > 0:
        print(f"Day {i+2}: Profit 📈")

    elif daily_change[i] < 0:
        print(f"Day {i+2}: Loss 📉")

    else:
        print(f"Day {i+2}: No Change")
```

### Best Buying Day

```python
buy_day = np.argmin(prices)

print("Best Buy Day:", buy_day + 1)

print("Price:", prices[buy_day])
```

### Best Selling Day

```python
sell_day = np.argmax(prices)

print("Best Sell Day:", sell_day + 1)

print("Price:", prices[sell_day])
```

### Overall Profit

```python
profit = prices[-1] - prices[0]

print("Overall Profit/Loss:")

if profit > 0:
    print("Profit:", profit)

elif profit < 0:
    print("Loss:", abs(profit))

else:
    print("No Profit No Loss")
```

### Percentage Return

```python
return_percentage = ((prices[-1] - prices[0]) / prices[0]) * 100

print("Return Percentage:")

print(round(return_percentage,2), "%")
```

### Volatility Analysis

```python
volatility = np.std(prices)

print("Market Volatility:\n")

if volatility > 10:
    print("High Volatility 🔥")

else:
    print("Low Volatility 😊")
```

### 3-Day Moving Average

```python
moving_average = np.convolve(
    prices,
    np.ones(3)/3,
    mode="valid"
)

print("3-Day Moving Average:\n")

print(moving_average)
```
