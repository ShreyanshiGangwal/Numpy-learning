# NumPy Banking System
---

```python
# Create Banking Data
import numpy as np

# Number of customers
n = int(input("Enter number of customers: "))

names = []
balances = []

# Input customer data
for i in range(n):

    name = input(f"\nEnter customer name {i+1}: ")

    balance = float(input(f"Enter balance for {name}: "))

    names.append(name)
    balances.append(balance)

# Convert balances into NumPy array
balances = np.array(balances)

print("\nCustomer Names:")
print(names)

print("\nBalances:")
print(balances)
```

### Check Total Bank Balance

```python
total_money = np.sum(balances)

print("\nTotal Money in Bank:", total_money)
```

### Find Richest Customer

```python
richest_index = np.argmax(balances)

print("\nRichest Customer:")
print(names[richest_index])

print("Balance:",
      balances[richest_index])
```

### Deposit Money

```python
customer = input("\nEnter customer name for deposit: ")

amount = float(input("Enter deposit amount: "))

if customer in names:
    index = names.index(customer)
    balances[index] += amount
    print("\nUpdated Balance:",
          balances[index])

else:
    print("Customer not found!")
```

### Withdraw Money

```python
customer = input("\nEnter customer name for withdrawal: ")

amount = float(input("Enter withdrawal amount: "))

if customer in names:

    index = names.index(customer)

    if balances[index] >= amount:

        balances[index] -= amount

        print("\nWithdrawal Successful")

        print("Remaining Balance:",
              balances[index])

    else:
        print("Insufficient Balance!")

else:
    print("Customer not found!")
```

### Low Balance Alert

```python
low_balance = balances[balances < 1000]

print("\nCustomers with Low Balance:")
print(low_balance)
```

### Transfer Money

```python
sender = input("\nEnter sender name: ")

receiver = input("Enter receiver name: ")

amount = float(input("Enter transfer amount: "))

if sender in names and receiver in names:

    sender_index = names.index(sender)
    receiver_index = names.index(receiver)

    if balances[sender_index] >= amount:

        balances[sender_index] -= amount
        balances[receiver_index] += amount

        print("\nTransfer Successful!")

    else:
        print("Insufficient Balance!")

else:
    print("Invalid Customer Name!")
```

### Display Final Bank Data

```python
print("\n----- FINAL BANK DATA -----")

for i in range(n):
    print(names[i],
          "-> ₹",
          balances[i])
```
