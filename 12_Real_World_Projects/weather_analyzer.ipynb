# Weather Data Analyzer
---

```python
import numpy as np

# Number of days
days = int(input("Enter number of days: "))

temperature = []

# Input temperatures
for i in range(days):

    temp = float(input(f"Enter temperature for Day {i+1}: "))

    temperature.append(temp)

# Convert to NumPy array
temperature = np.array(temperature)

print("\nTemperature Dataset:")
print(temperature)
```

### Basic Weather Analysis

```python
print("----- WEATHER ANALYSIS -----\n")

print("Average Temperature:",
      np.mean(temperature))

print("Maximum Temperature:",
      np.max(temperature))

print("Minimum Temperature:",
      np.min(temperature))

print("Total Temperature Sum:",
      np.sum(temperature))
```

### Find Hot & Cold Days

```python
# Hot days (>35)
hot_days = temperature[temperature > 35]

# Cold days (<20)
cold_days = temperature[temperature < 20]

print("Hot Days Temperatures:")
print(hot_days)

print("\nCold Days Temperatures:")
print(cold_days)
```

### Detect Rainy Days

```python
rainy_days = temperature[temperature < 25]

print("Possible Rainy Days:")
print(rainy_days)
```

### Celsius → Fahrenheit Conversion
- Formula

- F=(C × 5/9)+32

```python
fahrenheit = (temperature * 9/5) + 32

print("Temperature in Fahrenheit:")
print(fahrenheit)
```

### Weather Conditions Detector

```python
print("----- WEATHER CONDITIONS -----\n")

for temp in temperature:

    if temp >= 40:
        print(temp, "°C -> Very Hot")

    elif temp >= 30:
        print(temp, "°C -> Sunny")

    elif temp >= 20:
        print(temp, "°C -> Pleasant")

    else:
        print(temp, "°C -> Cold")
```

### Find Hottest Day

```python
hottest_day = np.argmax(temperature)

print("Hottest Day Index:",
      hottest_day)

print("Hottest Temperature:",
      temperature[hottest_day])
```

### Temperature Normalization

```python
normalized_temp = (
    (temperature - np.min(temperature))
    /
    (np.max(temperature) - np.min(temperature))
)

print("\nNormalized Temperature:")
print(normalized_temp)
```

### Weekly Weather Report

```python
print("\n----- WEEKLY WEATHER REPORT -----")

for i in range(days):
    print(f"Day {i+1} -> {temperature[i]} °C")
```
