# Cricket Score Analyzer
---

```python
import numpy as np
```

### Take Player Data

```python
n = int(input("Enter number of players: "))

players = []
runs = []

for i in range(n):

    name = input(f"Enter player {i+1} name: ")
    score = int(input(f"\nEnter runs scored by {name}: "))

    players.append(name)
    runs.append(score)

runs = np.array(runs)

print("Players:", players)
print("\nRuns:", runs)
```

### Team Statistics

```python
print("----- TEAM ANALYSIS -----\n")

print("Total Team Score:", np.sum(runs))

print("Average Score:", np.mean(runs))

print("Highest Score:", np.max(runs))

print("Lowest Score:", np.min(runs))
```

### Find Man of the Match

```python
top_scorer = np.argmax(runs)

print("Man of the Match:")

print(players[top_scorer])

print("Runs:", runs[top_scorer])
```

### Fifty & Century Detector

```python
print("----- SPECIAL ACHIEVEMENTS -----\n")

for i in range(len(runs)):

    if runs[i] >= 100:

        print(players[i], "scored a CENTURY")

    elif runs[i] >= 50:

        print(players[i], "scored a FIFTY")
```

### Performance Grades

```python
print("----- PLAYER GRADES -----\n")

for i in range(len(runs)):

    if runs[i] >= 100:
        grade = "A+"

    elif runs[i] >= 75:
        grade = "A"

    elif runs[i] >= 50:
        grade = "B"

    elif runs[i] >= 30:
        grade = "C"

    else:
        grade = "D"

    print(players[i], "->", grade)
```

### Above Average Players

```python
average = np.mean(runs)

print("Above Average Players:\n")

for i in range(len(runs)):

    if runs[i] > average:

        print(players[i], "-" , runs[i])
```

### Team Ranking

```python
sorted_scores = np.argsort(runs)[::-1]

print("----- PLAYER RANKING -----\n")

for rank, idx in enumerate(sorted_scores, start=1):

    print(rank , "." , players[idx] , "-" , runs[idx])
```

### Boundary Estimator
Assume: 
4 runs = boundary , 6 runs = six

```python
fours = runs // 4

print("Estimated Boundaries:\n")

for i in range(len(runs)):

    print(players[i], "≈", fours[i], "fours")
```

### Match Summary

```python
print("\n===== MATCH SUMMARY =====")

print("Total Runs:",
      np.sum(runs))

print("Average Runs:",
      round(np.mean(runs), 2))

print("Best Player:",
      players[np.argmax(runs)])

print("Best Score:",
      np.max(runs))
```
