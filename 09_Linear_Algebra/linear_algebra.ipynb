# Linear Algebra

```python
import numpy as np
```

 ---------------------------------------------------
### 1. Matrix multiplication

```python
A = np.array([
    [1, 2],
    [3, 4]
])

B = np.array([
    [5, 6],
    [7, 8]
])

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)

matrix_multi = np.matmul(A, B)

print("\nMatrix Multiplication:\n", matrix_multi)
```

 ---------------------------------------------------
### 2. Dot product

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

dot_product = np.dot(arr1, arr2)

print("Array 1:", arr1)
print("Array 2:", arr2)

print("\nDot Product:", dot_product)
```

 ---------------------------------------------------
### 3. Cross product

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

cross_product = np.cross(arr1, arr2)

print("Array 1:", arr1)
print("Array 2:", arr2)

print("\nCross Product:", cross_product)
```

 ---------------------------------------------------
### 4. Transpose matrix

```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

transpose = matrix.T

print("Matrix:\n", matrix)

print("\nTranspose Matrix:\n", transpose)
```

 ---------------------------------------------------
### 5. Find determinant
- Central Formula:
- det(A) = ad - bc

```python
matrix = np.array([
    [4, 6],
    [3, 8]
])

print("Matrix:\n", matrix)

determinant = np.linalg.det(matrix)

print("\nDeterminant:", determinant)

# Formula Visualization
# det(A) = ad - bc
# :contentReference[oaicite:0]{index=0}
```

 ---------------------------------------------------
### 6. Find inverse matrix

```python
matrix = np.array([
    [4, 7],
    [2, 6]
])

print('Matrix:\n', matrix)

inverse = np.linalg.inv(matrix)

print('\nInverse Matrix:\n', inverse)
```

 ---------------------------------------------------
### 7. Solve linear equations
- System representation:
- Ax = b

```python
a = np.array([
    [2, 1],
    [1, 3]
])

print('Matrix A:\n', a)

b = np.array([8, 13])

print('\nMatrix B:\n', b)

solve = np.linalg.solve(a, b)

print('\nSolution of Linear Equations:', solve)
```

 ---------------------------------------------------
### 8. Find eigenvalues & eigenvectors

```python
matrix = np.array([
    [4, 2],
    [1, 3]
])

print("Matrix:\n", matrix)

eigenvalues, eigenvectors = np.linalg.eig(matrix)

print("\nEigenvalues:\n", eigenvalues)

print("\nEigenvectors:\n", eigenvectors)
```

 ---------------------------------------------------
### 9. Calculate matrix rank

```python
matrix = np.array([
    [1, 2, 3],
    [2, 4, 6],
    [1, 1, 1]
])

print("Matrix:\n", matrix)

rank = np.linalg.matrix_rank(matrix)

print("\nMatrix Rank:", rank)
```

 ---------------------------------------------------
### 10. Build simple recommendation score matrix

```python
# Users vs Movies rating matrix
ratings = np.array([
    [5, 3, 0],
    [4, 0, 0],
    [1, 1, 0],
    [0, 0, 5]
])

print("Ratings Matrix:\n", ratings)

# Movie popularity score
recm_score = ratings.sum(axis=0)

print("\nRecommendation Scores:", recm_score)

print("\nMovie with Highest Score Index:",
      np.argmax(recm_score))
```
