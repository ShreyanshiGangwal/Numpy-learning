# Reshaping & Joining

```python
import numpy as np
```

 ---------------------------------------------------
### 1. Reshape: 1D → 2D and 2D → 3D

```python
# ---- 1D TO 2D ----
arr_1d = np.array([1, 2, 3, 4, 5, 6])

arr_2d = arr_1d.reshape(2, 3)

print("1D Array:", arr_1d)

print("\n1D to 2D:\n", arr_2d)

# ---- 2D TO 3D ----
arr_2d = np.array([
    [1, 2],
    [3, 4],
    [5, 6],
    [7, 8]
])

arr_3d = arr_2d.reshape(2, 2, 2)

print("\n2D Array:\n", arr_2d)

print("\n2D to 3D:\n", arr_3d)
```

 ---------------------------------------------------
### 2. Flatten matrix

```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

flattened = matrix.flatten()

print('Original Matrix:\n', matrix)
print('\nFlattened Matrix:\n', flattened)
```

 ---------------------------------------------------
### 3. Use: ravel() and flatten()

```python
matrix = np.array([
    [10, 20],
    [30, 40]
])

ravel_arr = matrix.ravel()
flatten_arr = matrix.flatten()

print("\nOriginal Matrix:")
print(matrix)

print("\nUsing ravel():", ravel_arr)

print("\nUsing flatten():", flatten_arr)
```

  ---------------------------------------------------
### 4. Stack arrays: horizontal and vertical

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

horizontal_stack = np.hstack((arr1, arr2))
vertical_stack = np.vstack((arr1, arr2))

print("Array 1:", arr1)
print("Array 2:", arr2)

print("\nHorizontal Stack:", horizontal_stack)

print("\nVertical Stack:\n", vertical_stack)
```

 ---------------------------------------------------
### 5. Split arrays equally.

```python
arr = np.array([1, 2, 3, 4, 5, 6])

equal_split = np.split(arr, 3)

print("Original Array:", arr)

print("\nEqual Split:", equal_split)
```

 ---------------------------------------------------
### 6. Split uneven arrays.

```python
arr = np.array([1, 2, 3, 4, 5, 6, 7])

uneven_split = np.array_split(arr, 3)

print("Original Array:", arr)

print("\nUneven Split:", uneven_split)
```

 ---------------------------------------------------
### 7. Concatenate arrays

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

concatenated = np.concatenate((arr1, arr2))

print("First Array:", arr1)
print("Second Array:", arr2)

print("\nConcatenated Array:", concatenated)
```

 ---------------------------------------------------
### 8. Repeat array elements.

```python
arr = np.array([1, 2, 3])

repeated = np.repeat(arr, 3)

print("Original Array:", arr)

print("\nRepeated Elements:", repeated)
```

 ---------------------------------------------------
### 9. Tile arrays.

```python
arr = np.array([1, 2, 3])

tiled = np.tile(arr, 3)

print("Original Array:", arr)

print("\nTiled Array:", tiled)
```

 ---------------------------------------------------
### 10. Swap rows and columns.

```python
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("Original Matrix:\n", matrix)

transposed = matrix.T

print("\nAfter Swapping Rows and Columns:\n", transposed)
```
