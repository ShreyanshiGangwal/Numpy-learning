# Student Result Analyzer
---

```python
import numpy as np

# Number of students
n = int(input("Enter number of students: "))

names = []
marks = []

# Taking input
for i in range(n):
    name = input(f"Enter name of student {i+1}: ")
    mark = int(input(f"Enter marks of {name}: "))

    names.append(name)
    marks.append(mark)

# Convert marks into NumPy array
marks = np.array(marks)

print("\nStudent Names:", names)
print("Student Marks:", marks)

# Basic statistics

print("\n----- RESULT ANALYSIS -----")

print("Average Marks:", np.mean(marks))

print("Highest Marks:", np.max(marks))

print("Lowest Marks:", np.min(marks))

# Topper
topper_index = np.argmax(marks)

print("Topper:", names[topper_index])

# Grade System
print("\n----- STUDENT GRADES -----")

for i in range(n):

    if marks[i] >= 90:
        grade = "A"
    elif marks[i] >= 75:
        grade = "B"
    elif marks[i] >= 60:
        grade = "C"
    elif marks[i] >= 40:
        grade = "D"
    else:
        grade = "F"

    print(names[i], "->", grade)

# Pass / Fail System
print("\n----- PASS / FAIL -----")

for i in range(n):
    if marks[i] >= 40:
        result = "PASS"
    else:
        result = "FAIL"

    print(names[i], "->", result)

# Percentage Calculation
print("\n----- PERCENTAGE -----")

for i in range(n):
    percentage = (marks[i] / 100) * 100

    print(names[i], "->", percentage, "%")
```
