# Array Operations

```python
import numpy as np
```

 ---------------------------------------------------
### 1. Perform: addition, subtraction, multiplication & division

```python
arr1 = np.array([10, 20, 30, 40, 50])
arr2 = np.array([1, 2, 3, 4, 5])

print('First Array is', arr1)
print('Second Array is', arr2)

add = arr1 + arr2
sub = arr1 - arr2
multiply = arr1 * arr2
div = arr1 / arr2

print("\nAddition:", add)
print("Subtraction:", sub)
print("Multiplication:", multiply)
print("Division:", div)
```

 ---------------------------------------------------
### 2. Square every element.

```python
arr = np.array([1, 2, 3, 4, 5])

sq = arr ** 2

print("Original Array:", arr)
print("\nSquare of Elements:", sq)
```

 ---------------------------------------------------
### 3. Find cube of elements.

```python
arr = np.array([1, 2, 3, 4, 5])

cube = arr ** 3

print("Original Array:", arr)
print("\nCube of Elements:", cube)
```

 ---------------------------------------------------
### 4. Find: max, min, mean, median, sum.

```python
arr = np.array([10, 20, 30, 40, 50])

print("Array:", arr)

print("\nMaximum:", np.max(arr))
print("Minimum:", np.min(arr))
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Sum:", np.sum(arr))
```

 ---------------------------------------------------
### 5. Add scalar value to array.

```python
arr = np.array([1, 2, 3, 4, 5])

add = arr + 10

print("Original Array:", arr)
print("\nAfter Adding Scalar:", add)
```

 ---------------------------------------------------
### 6. Multiply matrix with scalar.

```python
matrix = np.array([
    [1, 2],
    [3, 4]
])

result = matrix * 5

print("Original Matrix:\n", matrix)
print("\nAfter Scalar Multiplication:\n", result)
```

 ---------------------------------------------------
### 7. Compare two arrays.

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([1, 0, 3, 5])

print("First Array:", arr1)
print("Second Array:", arr2)

equal = arr1 == arr2
greater = arr1 > arr2
smaller = arr1 < arr2

print("\nEqual Comparison:", equal)
print("Greater Comparison:", greater)
print("Less Comparison:", smaller)
```

 ---------------------------------------------------
### 8. Find cumulative sum.

```python
arr = np.array([1, 2, 3, 4, 5])

cum_sum = np.cumsum(arr)

print("Array:", arr)
print("\nCumulative Sum:", cum_sum)
```

 ---------------------------------------------------
### 9. Find cumulative product.

```python
arr = np.array([1, 2, 3, 4, 5])

cum_product = np.cumprod(arr)

print("Array:", arr)
print("\nCumulative Product:", cum_product)
```

 ---------------------------------------------------
### 10. Normalize an array.
Formula:
- normalized_value = (x - min) / (max - min)

```python
arr = np.array([10, 20, 30, 40, 50])

norm = (arr - np.min(arr)) / (np.max(arr) - np.min(arr))

print('Original Array:', arr)
print('\nNormalized Array:', norm)
```
