# Mini Recommendation System
---

```python
import numpy as np
```

### Take User Ratings

```python
movies = np.array([
    "Avengers",
    "Until the End of the World",
    "Bahubali",
    "3 Idiots",
    "Inception"
])

ratings = []

print("Rate each movie from 1 to 5")

for movie in movies:
    rating = int(input(f"{movie}: "))
    ratings.append(rating)

ratings = np.array(ratings)

print("\nRatings:")
print(ratings)
```

### Average Rating

```python
print("Average Rating:")

print(np.mean(ratings))
```

### Favorite Movie

```python
best_movie = np.argmax(ratings)

print(f"The best movie is:\n{movies[best_movie]} with a rating of {ratings[best_movie]}")
```

### Least Favorite Movie

```python
least_movie = np.argmin(ratings)

print(f"The least movie is:\n{movies[least_movie]} with a rating of {ratings[least_movie]}")
```

### Recommend Movies

```python
recommendation = input("Do you want a recommendation? (yes/no): ")

if recommendation.lower() == "yes":

    max_rating = max(ratings)

    print("Top Recommended Movies:\n")

    for i in range(len(movies)):
        if ratings[i] == max_rating:
            print(f"☆ {movies[i]} (Rating: {ratings[i]})")

else:
    print("Thank you for using our recommendation system!")
```

### Sort Movies by Rating

```python
sorted_index = np.argsort(ratings)[::-1]

print("৹ Movie Ranking\n")

for rank, index in enumerate(sorted_index, start=1):

    print(f"{rank}. {movies[index]} - {ratings[index]} ☆")
```

### Rating Distribution

```python
unique, count = np.unique(
    ratings,
    return_counts=True
)

print("Rating Distribution\n")

for u, c in zip(unique, count):

    print(f"{u} Star : {c} Movie(s)")
```

### Find Movies Above Average

```python
avg = np.mean(ratings)

above_avg = movies[ratings > avg]

print("Movies with above average ratings:\n")

for r,m in enumerate(above_avg, start=1):
    print(f"{r}. {m}")
```

### Recommendation Score

```python
recommendation_score = ratings * 20

print("\nRecommendation Score")

for movie, score in zip(movies, recommendation_score):

    print(f'{movie} -> {score} %')
```

### Final Recommendation Report

```python
print("========== RECOMMENDATION REPORT ==========\n")

print("Average Rating:",
      round(np.mean(ratings),2))

print("\nFavorite Movie:",
      movies[np.argmax(ratings)])

print("Least Favorite:",
      movies[np.argmin(ratings)])

print("\nTop Recommended:")

for movie in movies[ratings >= 4]:

    print("☆", movie)
```
